```python
# Código para comparação entre os métodos intervalares da bissecção e da falsa posição
# Henrique Ferreira Mat.:251112972 Data:05/04/2025
import math
import matplotlib.pyplot as plt

#função para método da bissecção
def bisection(f, a, b, tol=1e-6, max_iter=1000):
    if f(a) * f(b) >= 0:
        raise ValueError("A função deve ter sinais opostos nos extremos do intervalo.")

    iter_count = 0
    errors = []
    roots = []
    c_prev = None

    while iter_count < max_iter:
        c = (a + b) / 2
        roots.append(c)

        if c_prev is not None:
            error = abs((c - c_prev) / c)
            errors.append(error)
            if error < tol:
                break
        else:
            errors.append(None)

        fc = f(c)

        if fc == 0:
            break

        if f(a) * fc < 0:
            b = c
        else:
            a = c

        c_prev = c
        iter_count += 1

    return roots, iter_count + 1, errors

#Função para método da falsa posição
def false_position(f, a, b, tol=1e-6, max_iter=1000):
    if f(a) * f(b) >= 0:
        raise ValueError("A função deve ter sinais opostos nos extremos do intervalo.")

    iter_count = 0
    errors = []
    roots = []
    c_prev = None

    while iter_count < max_iter:
        fa = f(a)
        fb = f(b)
        c = (a * fb - b * fa) / (fb - fa)
        roots.append(c)

        if c_prev is not None:
            error = abs((c - c_prev) / c)
            errors.append(error)
            if error < tol:
                break
        else:
            errors.append(None)

        fc = f(c)

        if fc == 0:
            break

        if fa * fc < 0:
            b = c
        else:
            a = c

        c_prev = c
        iter_count += 1

    return roots, iter_count + 1, errors

# Definindo as equações e intervalos
equations = [
    {
        'name': '-0.5x² + 2.5x + 4.5',
        'f': lambda x: -0.5*x**2 + 2.5*x + 4.5,
        'interval': (5, 10)
    },
    {
        'name': '5x³ -5x² +6x -2',
        'f': lambda x: 5*x**3 -5*x**2 +6*x -2,
        'interval': (0, 1)
    },
    {
        'name': '-25+82x-90x²+44x³-8x⁴+0.7x⁵',
        'f': lambda x: -25 +82*x -90*x**2 +44*x**3 -8*x**4 +0.7*x**5,
        'interval': (0.5, 1)
    },
    {
        'name': 'sin(x) - x³',
        'f': lambda x: math.sin(x) - x**3,
        'interval': (0.5, 1)
    },
    {
        'name': 'ln(x⁴) - 0.7',
        'f': lambda x: math.log(x**4) - 0.7,
        'interval': (0.5, 2)
    }
]

# Parâmetros numéricos
tolerance = 1e-6
max_iterations = 1000

# Processamento para cada equação
for eq in equations:
    print(f"\n{'='*50}")
    print(f"Resolvendo: {eq['name']}")
    print(f"Intervalo inicial: {eq['interval']}\n")

    f = eq['f']
    a, b = eq['interval']

    # Bissecção
    try:
        roots_bis, iter_bis, errors_bis = bisection(f, a, b, tolerance, max_iterations)
        print("Método da Bissecção:")
        print(f"Iterações: {iter_bis}")
        print(f"Raiz aproximada: {roots_bis[-1]:.8f}")
        print("Erro relativo por iteração:")
        for i, err in enumerate(errors_bis):
            print(f"Iteração {i+1}: {err if err is not None else 'N/A'}")
    except ValueError as e:
        print(f"Erro na bissecção: {e}")

    # Falsa Posição
    try:
        roots_fp, iter_fp, errors_fp = false_position(f, a, b, tolerance, max_iterations)
        print("\nMétodo da Falsa Posição:")
        print(f"Iterações: {iter_fp}")
        print(f"Raiz aproximada: {roots_fp[-1]:.8f}")
        print("Erro relativo por iteração:")
        for i, err in enumerate(errors_fp):
            print(f"Iteração {i+1}: {err if err is not None else 'N/A'}")
    except ValueError as e:
        print(f"Erro na falsa posição: {e}")

    # Plotagem comparativa
    plt.figure(figsize=(10, 6))
    valid_bis = [e for e in errors_bis if e is not None]
    valid_fp = [e for e in errors_fp if e is not None]

    plt.plot(range(1, len(valid_bis)+1), valid_bis, 'b-o', label='Bissecção')
    plt.plot(range(1, len(valid_fp)+1), valid_fp, 'r--s', label='Falsa Posição')

    plt.xlabel('Iteração')
    plt.ylabel('Erro Relativo')
    plt.title(f'Comparação de Convergência para: {eq["name"]}')
    plt.yscale('log')
    plt.grid(True)
    plt.legend()
    plt.show()
```
